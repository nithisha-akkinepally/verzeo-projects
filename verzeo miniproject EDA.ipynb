```python
data['total marks']=data['math score']+data['reading score']+data['writing score']
data['percentage']=data['total marks']/300*100
```

```python
#dividing data on basis of grades
```

```python
def grade(score):
    if score>=85 and score<=100:
        return 'Grade A'
    elif score>=70 and score<=84:
        return 'Grade B'
    elif score>=55 and score<=69:
        return 'Grade C'
    elif score>=35 and score<=54:
        return 'Grade D'
    else:
        return 'Grade E'
    
```

```python
Grade=[grade(i) for i in data['percentage']]
data.insert(10,'grade',Grade,True)
```

```python
data['grade'].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("grades pie chart",fontsize=20)
```

```python
print("Bar graph -grades")
gr=pd.crosstab(data['grade'],data['race/ethnicity'],normalize=0)
gr.plot.bar(stacked=True)
gr=pd.crosstab(data['grade'],data['parental level of education'],normalize=0)
gr.plot.bar(stacked=True)
gr=pd.crosstab(data['grade'],data['test preparation course'],normalize=0)
gr.plot.bar(stacked=True)
gr=pd.crosstab(data['grade'],data['lunch'],normalize=0)
gr.plot.bar(stacked=True)
```
